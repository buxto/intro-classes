```python
class Animal:
    """Animal Class"""
    def __init__(self, sci_name, pettable, alive_status):
        self._sci_name = sci_name
        self._pettable = pettable
        self._alive_status = alive_status
    
    @property
    def sci_name(self):
        return self._sci_name
    
    @property
    def pettable(self):
        return self._pettable
    
    @property
    def alive_status(self):
        return self._alive_status
    
    @sci_name.setter
    def sci_name(self, sci_name):
        self._sci_name = sci_name
        
    @pettable.setter
    def pettable(self, pettable):
        self._pettable = pettable
    
    @alive_status.setter
    def alive_status(self, alive_status):
        self._alive_status = alive_status
        
    def move(self):
        print('Moving')
        
    def eat(self):
        print('Eating')
        
    def reproduce(self):
        print('Reproducing')
    
    def sleep(self):
        print('ZZZ')
    
    def die(self):
        print('This creature has died')
        pettable(self, False)
    
```

```python
class Book:
    def __init__(self, title, author, edition, coauthors, date, original_pub_date, publisher, genre):
        self._title = title
        self._author = author
        self._edition = edition
        self._coauthors = coauthors
        self._date = date
        self._original_pub_date = original_pub_date
        self._publisher = publisher
        self._genre = genre
        
    @property
    def title(self):
        return self._title
    @property
    def author(self):
        return self._author
    @property
    def edition(self):
        return self._edition
    @property
    def coauthors(self):
        return self._coauthors
    @property
    def date(self):
        return self._date
    @property
    def original_pub_date(self):
        return self._original_pub_date
    @property
    def publisher(self):
        return self._publisher
    @property
    def genre(self):
        return self._genre
```

```python
class Vehicle:
    def __init__(self, model, manufacturer, year, engine_type, needs_repairs = False):
        self._model = model
        self._manufacturer = manufacturer
        self._year = year
        self._engine_type = engine_type
        self._needs_repairs = needs_repairs
        
    @property
    def model(self):
        return self._model
    @property
    def manufacturer(self):
        return self._manufacturer
    @property
    def year(self):
        return self._year
    @property
    def engine_type(self):
        return self._engine_type
    @property
    def needs_repairs(self):
        return self._needs_repairs
    @needs_repairs.setter
    def needs_repairs(self, status):
        self._needs_repairs = status
    
    def drive(self):
        print('Moving')
    def park(self):
        print('Parking')
    def crash(self):
        print('This vehicle has crashed')
        needs_repairs(self, True)
    def repair(self):
        print('This vehicle is repaired')
        needs_repairs(self, False)
        
```

```python
class Fish(Animal):
    def __init__(self, sci_name, pettable, alive_status, water_type, habitat_zone, habitat_region, eats_type):
        super.__init__(sci_name, pettable, alive_status)
        self._water_type = water_type
        self._habitat_zone = habitat_zone
        self._habitat_region = habitat_region
        self._eats_type = eats_type
    @property
    def water_type(self):
        return self._water_type
    @property
    def habitat_zone(self):
        return self._habitat_zone
    @property
    def habitat_region(self):
        return self._habitat_region
    @property
    def eats_type(self):
        return self._eats_type
    
    def swim(self):
        print('Swimming')
    def glub(self):
        print('Glubglub')
```

```python
class Snake(Animal):
    def __init__(self, sci_name, pettable, alive_status, habitat_zone, habitat_region, eats_type, venomous):
        super.__init__(sci_name, pettable, alive_status)
        self._habitat_zone = habitat_zone
        self._habitat_region = habitat_region
        self._eats_type = eats_type
        self._venomous = venomous
    @property
    def habitat_zone(self):
        return self._habitat_zone
    @property
    def habitat_region(self):
        return self._habitat_region
    @property
    def eats_type(self):
        return self._eats_type
    @property
    def venomous(self):
        return self._venomous
    
    def bite(self):
        print('The snake bites')
```

```python
class Person(Animal):
    def __init__(self, sci_name, pettable, alive_status, person_name, person_location):
        super.__init__(sci_name, pettable, alive_status)
        self._person_name = person_name
        self._person_location = person_location
    @property
    def person_name(self):
        return self._person_name
    @property
    def person_location(self):
        return self._person_location
    @person_name.setter
    def person_name(self, name):
        self._person_name = name
    @person_location.setter
    def person_location(self, location):
        self._person_location = location
```

```python
class Textbook(Book):
    def __init__(self, title, author, edition, coauthors, date, original_pub_date, publisher, genre, subjects, condition):
        super.__init__(title, author, edition, coauthors, date, original_pub_date, publisher, genre)
        self._subjects = subjects
        self._condition = condition
    @property
    def subjects(self):
        return self._subjects
    @property
    def condition(self):
        return self._condition
    @subjects.setter
    def subjects(self, subjects):
        self._subjects = subjects
    @condition.setter
    def condition(self, condition):
        self._condition = condition
```

```python
class AddressBook(Book):
    def __init__(self, title, author, edition, coauthors, date, original_pub_date, publisher, genre, addresses, phone_num, name):
        super.__init__(title, author, edition, coauthors, date, original_pub_date, publisher, genre)
        self._addresses = addresses
        self._phone_num = phone_num
        self._name = name
    @property
    def addresses(self):
        return self._addresses
    @property
    def phone_num(self):
        return self._phone_num
    @property
    def name(self):
        return self._name
    @addresses.setter
    def addresses(self, addresses):
        self._addresses = addresses
    @phone_num.setter
    def phone_num(self, phone_num):
        self._phone_num = phone_num
    @name.setter
    def name(self, name):
        self._name = name
```
